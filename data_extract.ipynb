```python
def data_from_array(array_data):
    new_data = []
    for item in array_data:
        if item != "":
            new_data.append(item)
    return new_data
```

```python
def list_to_string(array_list):
    string = ''
    for num in range(1, len(array_list)):
        if num == 1: 
            string += str(array_list[num])
        elif num % 2 != 0 and num != 1: 
            string += ";" + str(array_list[num])
        else:
            string += "," + str(array_list[num])
    return string
```

```python
# Okay, now we have gotten all the keys out.
# Ah, now I need to read the files. 
items = set()
with open("keys.txt", "r") as f:
    while True:
        string = f.readline().strip()
        items.add(string)
        if not string:
            break 
print(items)
data_new = {}
for item in items:
    data_new[item] = []
data_new['Bus Stops'] = []
data_new["Primary Schools"] = []
data_new["Groceries & Supermarts"] = []
data_new["Shopping Malls"] = []
data_new["Trains (MRT & LRT)"] = []
print(data_new)

for key in data_new:
    for num in range(0, 4252):
        data_new[key].append("NA")

print(data_new)
for keys in data_new:   
    print(len(data_new[keys]))
```

```python
items = set()
for num in range(0, 4252):
    soup = listing_link(num+1)
    about_this_property = soup.findAll("div", class_="row listing-about")
    date_listed = soup.findAll("div", class_="row listing-about d-none")
    nearby_train = soup.findAll("div", class_="col-12 col-md-6 col-lg-4 list-of-amenity Trains")
    nearby_bus_stop = soup.findAll("div", class_="col-12 col-md-6 col-lg-4 list-of-amenity Bus-Stops")
    nearby_schools = soup.findAll("div", class_="col-12 col-md-6 col-lg-4 list-of-amenity Schools")
    nearby_shopping = soup.findAll("div", class_="col-12 col-md-6 col-lg-4 list-of-amenity Shopping-Malls")
    nearby_markets = soup.findAll("div", class_="col-12 col-md-6 col-lg-4 list-of-amenity Markets")
    # Now comes the data_new.
    # data_new[""] = num
    for number in range(0, len(about_this_property)):
        test = data_from_array(about_this_property[number].text.strip("\n\t ").split("\n"))
        items.add(about_this_property[number].text.strip("\n\t ").split("\n")[0])
        data_new[about_this_property[number].text.strip("\n\t ").split("\n")[0]][num] = (list_to_string(test))
    if date_listed:
        date = data_from_array(date_listed[0].text.strip("\n\t").split("\n"))
        data_new["Date Listed"][num] = (list_to_string(date))
    if nearby_bus_stop:
        bus_stop = data_from_array(nearby_bus_stop[0].text.strip("\n\t").split("\n"))
        data_new['Bus Stops'][num] = (list_to_string(bus_stop))
    if nearby_train:
        train = data_from_array(nearby_train[0].text.strip("\n\t").split("\n"))
        data_new["Trains (MRT & LRT)"][num] = (list_to_string(train))
    if nearby_schools:
        schools = data_from_array(nearby_schools[0].text.strip("\n\t").split("\n"))
        data_new["Primary Schools"][num] = (list_to_string(schools))
    if nearby_shopping:
        shopping = data_from_array(nearby_shopping[0].text.strip("\n\t").split("\n"))
        data_new["Shopping Malls"][num] = (list_to_string(shopping))
    if nearby_markets:
        markets = data_from_array(nearby_markets[0].text.strip("\n\t").split("\n"))
        data_new["Groceries & Supermarts"][num] = (list_to_string(markets))

print(items)
```
