```python
df_addr = pd.DataFrame(unique_addr, columns=['Full_Addr'])
df_addr
```

```python
df_addr.isna().sum()
```

```python
from time import sleep
from random import randint
from geopy.geocoders import Nominatim
from geopy.exc import GeocoderTimedOut, GeocoderServiceError

user_agent = 'user_me_{}'.format(randint(10000,99999))
geolocator = Nominatim(user_agent=user_agent, timeout=5)
def reverse_geocode(address, sleep_sec=1):
    try:
      location = geolocator.geocode(address)  
      print(location)
      if location:
        return location.latitude, location.longitude
      else:
        return None, None
    except GeocoderTimedOut:
        print('TIMED OUT: GeocoderTimedOut: Retrying...')
        sleep(randint(1*100,sleep_sec*100)/100)
        return reverse_geocode(address,sleep_sec)
    except GeocoderServiceError as e:
        print('CONNECTION REFUSED: GeocoderServiceError encountered.')
        return None, None
    except Exception as e:
        print('ERROR: Terminating due to exception {}'.format(e))
        return None, None


for index, row in df_addr.iterrows():
  lat, longi = reverse_geocode(row['Full_Addr'])
  df_addr.loc[index, 'latitude'] = lat
  df_addr.loc[index, 'longitude'] = longi

df_addr.to_csv('data_with_coordinates.csv', index=False)
```

```python
import folium

def plotDot(latitude, longitude, _map, color, tip=''):
  folium.CircleMarker(location=[latitude, longitude],
                      radius=3,
                      weight=10,
                      popup=tip,
                      color=color).add_to(_map)
```
