# EDA: Geographical Analysis

```python
# Remove NaN values for coords for the purpose of Geo analysis
df_condo_filtered = merged_df.dropna(subset=['nearest_station_lat'])
```

```python
import folium

def createMap():
  return folium.Map(location=[1.3521, 103.8198])

def plotDot(latitude, longitude, _map, color, tip=''):
  folium.CircleMarker(location=[latitude, longitude],
                      radius=3,
                      weight=3,
                      popup=tip,
                      color=color).add_to(_map)

def plotIcon(latitude, longitude, _map, color, tip=''):
  folium.Marker(location=[latitude, longitude],
                      icon=folium.Icon(color=color),
                      popup=tip).add_to(_map)

# Map centralised in Singapore
condo_map = createMap()
```

### Plot all the train stations

```python
for index, row in df_train_coords.iterrows():
    plotIcon(latitude=row['lat'], longitude=row['lng'], _map=condo_map, color='blue', tip=row['station_name'])

condo_map
```

### Heatmap for Number of Condo listings

```python
from folium.plugins import HeatMap as heat_map

heatmap_condo = createMap()
gradient = {0.2: 'blue', 0.4: 'green', 0.6: 'orange', 1: 'red'}

heatmap_condo_data = df_condo_filtered[['nearest_station_lat', 'nearest_station_lng']].values.tolist()
heatmap_condo.add_child(heat_map(heatmap_condo_data, radius=15, opacity=1, min_opacity=0.05, gradient=gradient))
heatmap_condo
```

### Heatmap based on the price of condo listing

```python
# Get average of condo prices grouped by train stations
df_avg = df_condo_filtered.groupby(['nearest_station_lat', 'nearest_station_lng'])['Asking'].mean().reset_index()
df_avg = pd.DataFrame(df_avg)
df_avg = pd.merge(df_avg, df_train_coords, left_on=['nearest_station_lat', 'nearest_station_lng'], right_on=['lat', 'lng'], how='left')
df_avg.drop(['type', 'lat', 'lng'], axis=1, inplace=True)
df_avg.sort_values(by='Asking', ascending=False).head(10)
```
