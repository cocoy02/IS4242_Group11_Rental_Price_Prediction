# Preprocessing

## Join Review and Condo

```python
merged_df = pd.merge(df_rent, df_review, how="left", left_on="Property Name", right_on="Condominium Name")
merged_df
```

## Drop columns

```python
merged_df.drop(['Developer', 'Property', 'Address', 'Date Listed', 'Condominium Name'], axis=1, inplace=True)
```

## Drop Missing Values

```python
merged_df.isna().sum()
```

```python
merged_df.dropna(inplace=True)
```

## One Hot Encoding

```python
from datetime import datetime
cat_vars = ['Lease Term', 'District', 'Room Type', 'Trains (MRT & LRT)', 'Neighborhood']

# convert all categorical variables
for cat_var in cat_vars:
  merged_df = pd.concat([merged_df, pd.get_dummies(merged_df[cat_var])], axis=1)
```

```python
# drop org columns after one hot encoding
merged_df.drop(cat_vars, axis=1, inplace=True)
```

## Clean Feature Names

```python
import re

# replace special characters with underscores --> cos of the 2 years vs 2+ years
merged_df.columns = [re.sub(r'[^\w\s]+', '_', c) for c in merged_df.columns]

# replace space with underscore
merged_df.columns = merged_df.columns.str.replace(' ', '_')
```

# Get Model

```python
import pickle
with open('/content/drive/MyDrive/IS4242/rf_model_reviews.pkl', 'rb') as f:
    model = pickle.load(f)
```

```python
import math
def predict_price(df_new):
  price = model.predict(df_new)
  return math.exp(price[0]) #bc price is in log
```

# Convert Inputs



## 1. Get Other Feature Values from Property Name

```python
def get_features_from_name(property_name):
  row = merged_df.loc[merged_df['Property_Name'] == property_name]
  first_row = row.iloc[0]

  # Create test dataframe
  df_new = pd.DataFrame(first_row).transpose()
  df_new.drop(['Asking', 'Property_Name'], axis=1, inplace=True)

  return df_new
```

## 2. Scale and add input values to dataframe

```python
# Scale inputs before adding to dataframe
from sklearn.preprocessing import MinMaxScaler
from sklearn.compose import ColumnTransformer
import joblib

def scale_numeric_inputs(bathrooms, bedrooms):
  scaler = joblib.load(open('/content/drive/MyDrive/IS4242/Price Data Preprocess_1.2.2ver.sav', 'rb'))
  new_data = pd.DataFrame({'Bathrooms': [bathrooms], 'Bedrooms': [bedrooms]})
  transformed_data = scaler.transform(new_data)
  print(transformed_data)
  return transformed_data[0][0], transformed_data[0][1]
```

```python
def add_input_to_df(df_new, lease_term, room_type, bedrooms, bathrooms):
  # convert lease_term and room type column names 
  lease_term = lease_term.replace(' ', '_')
  lease_term = re.sub(r'[^\w\s]+', '_', lease_term)
  room_type = room_type.replace(' ', '_')

  df_new['Bathrooms'] = bathrooms
  df_new['Bedrooms'] = bathrooms
  df_new[lease_term] = 1
  df_new[room_type] = 1
```

```python
def scale_add_inputs(df_new, lease_term, room_type, bedrooms, bathrooms):
  scaled_bathrooms, scaled_bedrooms = scale_numeric_inputs(bathrooms, bedrooms)
  add_input_to_df(df_new, lease_term, room_type, scaled_bedrooms, scaled_bathrooms)
```

```python
import gradio as gr
from gradio.blocks import *
```

```python
def predict_rental_price(property_name, lease_term, room_type, bedrooms, bathrooms):
  print(property_name, lease_term, room_type, bedrooms, bathrooms)
  # 1.Get other values from property_name
  df_new = get_features_from_name(property_name)
  # 2. Scale and Add user inputs into dataframe
  scale_add_inputs(df_new, lease_term, room_type, bedrooms, bathrooms)
  # 3. Predict using model
  price = predict_price(df_new)
  return round(price) 

def get_review_analytics(property_name):
  row = df_review[df_review['Condominium Name'] == property_name]
  row = row.iloc[0]
  location_rating = row['Location']
  facilities_rating = row['Facilities']
  fin_rating = row['Financial Value']
  sm_rating = row['Service and Management']
  env_rating = row['Environment']
  return round(location_rating, 3), round(facilities_rating, 3), round(fin_rating, 3), round(sm_rating, 3), round(env_rating, 3)

with gr.Blocks(css=".gradio-container {width: 100%} .tab-label {font-weight: bold, font-size: 16px}") as demo:
  with gr.Row():
    gr.Markdown('# A Better Model 🏡')
  gr.HTML(
  "<div style='background-image: url(\"file/drive/MyDrive/IS4242/condobg2.png\"); background-size: cover; background-position: center; height: 40vh; z-index:-1'></div>\
    <div style='position: absolute;top: 0;left: 0;width: 100%;height: 100%; z-index=1'></div>\
    <div style='position: absolute;top: 0;left: 0;width: 100%;height: 100%;background-color: rgba(255, 255, 255, 0.5); z-index: 1'></div>\
    <div style='position: absolute; top: 50%; left: 50%; transform: translate(-50%, -50%); text-align: center;z-index: 2'>\
      <p style='font-size: 38px; font-weight: bold; color: #ffffff'>Predict Condo Prices ft. Review Analytics</p>\
    </div>"
  )

  with gr.Tab("Predict Prices 🤑", elem_classes='tab-label'):
    with gr.Row():
      with gr.Column(): # cat
        property_name = gr.Dropdown(property_name_list, label = "Property Name")
        lease_term = gr.Dropdown(lease_term_list, label='Lease Term')
        room_type = gr.Dropdown(room_type_list, label= "Room Type")

      with gr.Column(): # numeric
        bedrooms = gr.Number(label = "Bedrooms")
        bathrooms = gr.Number(label = "Bathrooms")
    with gr.Row():
      predict_price_btn = gr.Button("Predict", variant="primary")
    with gr.Row():
      output = gr.Textbox(label="Output")
    predict_price_btn.click(fn=predict_rental_price, inputs=[property_name, lease_term, room_type, bedrooms, bathrooms], outputs=output)
  
  with gr.Tab("Get Review Analysis 🔍", elem_classes='tab-label'):
    with gr.Row():
      with gr.Column():
        condo_name = gr.Dropdown(property_name_list, label = "Property Name")
        get_reviews_btn = gr.Button('Get Review Analytics', variant="primary")
      with gr.Column():
        output_location = gr.Textbox(label="Location Score (on a scale of 0-1)")
        output_facilities = gr.Textbox(label="Facilities Score (on a scale of 0-1)")
        output_fin = gr.Textbox(label="Financial Value Score (on a scale of 0-1)")
        output_sm = gr.Textbox(label="Service & Management Score (on a scale of 0-1)")
        output_env = gr.Textbox(label="Environment Score (on a scale of 0-1)")
    get_reviews_btn.click(fn=get_review_analytics, inputs=[condo_name], outputs=[output_location, output_facilities, output_fin, output_sm, output_env])    

demo.launch(debug=True, show_api=False)
```
