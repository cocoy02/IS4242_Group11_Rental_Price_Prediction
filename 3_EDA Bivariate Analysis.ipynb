```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
condo = pd.read_csv("df_srx_Condo.csv", index_col=0)
```

# Condo

```python
pairplot = sns.pairplot(data=condo)
pairplot.fig.set_size_inches(15, 15)
pairplot.fig.suptitle("Pair Plot", y=1.02)
plt.show()
```

**Findings**

- The condos built in recent years, the amenities such as shopping malls and schools are closer.
- Size of the condo is getting smaller and smaller

```python
plt.figure(figsize=(20,10))
sns.boxplot(x = "District",y="Size", data=condo)
```

```python
plt.figure(figsize=(20,10))
sns.boxplot(x = "District",y="Asking", data=condo)
```

**Findings**

- D10, D1, D4, D11 has larger size and higher rental price since in central area or places always crowded.

- D18, D26 smaller size and lower rental price

```python
fig = plt.figure(figsize=(20,10))
fig.add_subplot(1, 2, 1)   
sns.boxplot(x = "Bathrooms",y="Asking", data=condo)
fig.add_subplot(1, 2, 2) 
sns.boxplot(x = "Bathrooms",y="Size", data=condo)
```

```python
fig = plt.figure(figsize=(20,10))
fig.add_subplot(1, 2, 1)   #top and bottom left
sns.boxplot(x = "Bedrooms",y="Asking", data=condo)
fig.add_subplot(1, 2, 2)   #top right
sns.boxplot(x = "Bedrooms",y="Size", data=condo)
```

**Findings**

- Bathrooms and Bedrooms has positive relationship with Size and Rental Price

```python
fig = plt.figure(figsize=(20,10))
fig.add_subplot(1, 2, 1)   #top and bottom left
sns.boxplot(x = "Neighborhood",y="Asking", data=condo)
fig.add_subplot(1, 2, 2)   #top right
sns.boxplot(x = "Neighborhood",y="Size", data=condo)
```

**Findings**

- In Condonomium, smaller neighborhood could have larger size and higher price since more safe and privacy

```python
plt.figure(figsize=(10,20))
sns.boxplot(y="Trains (MRT & LRT)", x="Asking", data=condo)
```

```python
plt.figure(figsize=(10,20))
sns.boxplot(y="Trains (MRT & LRT)", x="Size", data=condo)
```

**Findings**

- Near central area, such as Orchard MRT and Thomson MRT, size and price is larger
